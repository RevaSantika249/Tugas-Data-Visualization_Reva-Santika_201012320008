```python
import io
df = pd.read_excel(io.BytesIO(uploaded['epl-goalScorer(20-21).xlsx']))
df
```

```python
# Nomor 1 , 10 player terbanyak mencetak gol
df_desc = df.sort_values(by='goals', ascending=False)
df_desc[['player_name','goals']].head(10)
```

```python
# Nomor 2 , 10 player terlama durasi bermain
df_desc = df.sort_values(by='time', ascending=False)
df_desc[['player_name','time']].head(10)
```

```python
# Nomor 3 , 10 player terbanyak assist
df_desc = df.sort_values(by='assists', ascending=False)
df_desc[['player_name','assists']].head(10)
```

```python
# Nomor 4 , 4 player pencetak gol terbanyak dari kesebelasan manchester city
df_mancity = df[df.team_title=='Manchester City']
df[['player_name','goals']].head(4)
```

```python
# Nomor 5 , Dari kesebelasan mana pencetak gol terbanyak?
df_lotgoals = df.sort_values(by='goals', ascending=False)
df[['team_title','goals']].head(1)
```

```python
# Nomor 6 , Berapa jumlah kartu kuning dan merah yang diterima pembuat assist terbanyak?
df0=df['yellow_cards']+df['red_cards']
df1=df.sort_values(by='assists', ascending=False)
df2=df1[['player_name','assists','yellow_cards','red_cards']].assign(yellow_red=df0)
df2.head(1)
```

```python
# Nomor 7 , Siapa yang jumlah assist dan golnya paling banyak dan dari kesebelasan mana?
df0=df['goals']+df['assists']
df1=df[['player_name','team_title','assists','goals']].assign(goals_assists=df0)
df1.head(1)
```

```python
# Nomor 8 , 5 kesebelasan yang menyumbangkan gol terbanyak
df_lotgoals=df.groupby('team_title').sum().sort_values(by='goals',ascending=False)
df[['goals']].head(5)
```

```python
# Nomor 9 , Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali!
df0=df['yellow_cards']+df['red_cards']
df1=df[['player_name','goals','yellow_cards','red_cards']].assign(yellow_red=df0)
df2=df1[df1.yellow_red==0].sort_values(by='goals',ascending=False)
df2.head(10)
```
